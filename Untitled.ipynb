# LATIN HYPERCUBE SAMPLING

```python
def LHS(n, p, sx=0, dx=0):
    
    sx = numpy.arange(0, n)/n
    dx = numpy.arange(1, n+1)/n
    
    punti = numpy.random.uniform(low = sx, high= dx, size=[p,n]).T
    
    for j in range(p):
        numpy.random.shuffle(punti[:, j])
    
    return punti
```
